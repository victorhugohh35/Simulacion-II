# *Girasol* #


```python
import math
import turtle
from turtle import * 

turtle.bgcolor("black")
turtle.shape("turtle")
turtle.speed(0)
turtle.goto(0, -40)
turtle.pendown()

for i in range(10):
    for j in range(18):
        turtle.color("yellow","brown")
        turtle.rt(90)
        turtle.circle(150 - j * 6, 90)
        turtle.lt(90)
        turtle.circle(150 - j * 6, 90)
        turtle.rt(180)
        turtle.circle(40, 24)


turtle.pendown()
turtle.color("#8B4513")
penup()
goto(0, 0)
pendown()
cont = 1
while cont < 130:
    forward(cont*0.2)
    right(320)
    cont += 1
turtle.hideturtle()
turtle.done()
```

# *Espiral* #


```python
import turtle
turtle.speed(0)
turtle.bgcolor("lightblue")
#turtle.bgcolor("green")
turtle.shape("triangle")
turtle.shapesize(0.5, 0.5)
turtle.color("black")
#turtle.color("pink")
for i in range(100):
    forward(i)
    left(30)
    turtle.stamp()
turtle.hideturtle()
done()
```

# *Poligono de n lados* #


```python
import turtle
import random
turtle.home()
lados = 12
tam = 70
angulo = 360.0 / lados

# Rellenar la figura con un color aleatorio
turtle.color(random.random(), random.random(), random.random())

turtle.begin_fill()
for i in range(lados): 
    turtle.forward(tam)
    turtle.right(angulo)
turtle.end_fill()
turtle.hideturtle()      
turtle.done() 
```

# *Carita Feliz* #


```python
import turtle
from turtle import * 

t = turtle.Turtle()
t.speed(0)
t.hideturtle()

rojo = "#FF0000"
azul = "#0000FF"
blanco = "#FFFFFF"
negro = "#000000"

t.penup()
t.goto(0, -150)
t.pendown()
t.begin_fill()
t.color(rojo)
t.circle(150)
t.end_fill()

# Dibujar los ojos
t.penup()
t.goto(-60, 0)
t.pendown()
t.begin_fill()
t.color(blanco)
t.circle(60)
t.end_fill()

t.penup()
t.goto(60, 0)
t.pendown()
t.begin_fill()
t.color(blanco)
t.circle(60)
t.end_fill()

t.penup()
t.goto(-50, 0)
t.pendown()
t.begin_fill()
t.color(negro)
t.circle(50)
t.end_fill()

t.penup()
t.goto(70, 0)
t.pendown()
t.begin_fill()
t.color(negro)
t.circle(50)
t.end_fill()

t.penup()
t.goto(-60, -50)
t.pendown()
t.width(8)
t.color(negro)
t.right(45)
t.circle(85, 90)

t.hideturtle()
turtle.done()
```
