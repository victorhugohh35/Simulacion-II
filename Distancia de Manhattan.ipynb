# Distancia Manhattan
## Caminata 2D

Consideremos que una persona se puede mover al azar en cualquiera de las 4 direcciones {N, S, E, O}.
Despues de cierto numeros de pasos, la persona estara en la posición $(x_f, y_f$, es decir, la persona esta a una **distancia Manhattan** del punto inicial $(x_i,y_2)$ o l2 igual a:
\begin{equation*}
d_M = |x_f - x_i| + |y_f - y_i|
\end{equation*}
Necesitamos, determinar cual es la **longitud (pasos)** que requiere la persona para que la probabilidad de que despues de haber dado esos **N** pasos se enucnetre a una distancia Manhattan no más de 4 sea del 50%

```python
import random
from matplotlib import pyplot as plt
xi = 0
yi = 0
movx = [1, -1] 
movy = [1, -1]

def distanciaM(x, y):
    d = abs(x-xi) + abs(y-yi)
    return d
    
def caminata(n):
    x = 0
    y = 0
    for i in range(1,n+1):
        if random.random() < 0.5:
            x = x + random.choice(movx)
        else:
            y = y + random.choice(movy)
    return x, y

#Esta lista guarda la prob simulada para cada n de 1 a 100
l = []
simulaciones=100000
for j in range(1,100+1): #for va desde 1 hasta n-1 for j in range(1,n)
    aciertos=0
    for m in range(1,simulaciones+1):
        a, b = caminata(j)
        if distanciaM(a, b)<=4:
            aciertos += 1
    l.append(aciertos/simulaciones)
        

```

```python
plt.plot(l)
plt.show()
```

```python
def encontrar_mas_cercano_a_0_5(lista):
    # Inicializamos la distancia mínima con un valor grande
    distancia_minima = float('inf')
    # Inicializamos la posición del valor más cercano
    posicion_mas_cercana = None
    
    # Iteramos sobre los elementos de la lista
    for i, valor in enumerate(lista):
        # Calculamos la distancia absoluta entre el valor actual y 0.5
        distancia_actual = abs(valor - 0.5)
        # Si la distancia actual es menor que la mínima encontrada hasta ahora
        if distancia_actual < distancia_minima:
            # Actualizamos la distancia mínima y la posición del valor más cercano
            distancia_minima = distancia_actual
            posicion_mas_cercana = i
    
    # Devolvemos la posición del valor más cercano a 0.5
    return posicion_mas_cercana

# Ejemplo de uso
posicion = encontrar_mas_cercano_a_0_5(l)
print("El valor más cercano a 0.5 está en la posición:", posicion)
```

```python
indice = 100
for i in range(len(l)):
    if l[len(l)-1-i] >= 0.5:
        break
    indice-=1
print("El primer valor más grande o igual a 0.5 es:",indice)
```

# ¿En que posición estara en N pasos?, $E[X_t]$
