## Ejemplo

Calcule la probabilidad de que aparezcan dos caras lanzando tres veces una moneda al aire.

1. Analíticamente.
2. Mediante simulación.

Analiticamente:

El tiempo de llegada de cualquier persona, digamos $x_1,\ x_2$ está representada por una variable aleatoria uniformemente distribuida en $[0,30]$, es decir.
$$X_1, X_2 \sim U(0,30)$$

Ahora, asumiendo independencia entre estas variables, tienen por densidad marginal
\begin{equation*}
\begin{aligned}
f_{X_1} (x_1) = \frac{1}{30}I_{(0,30)}(x_1) && f_{X_2} (x_2) = \frac{1}{30}I_{(0,30)}(x_2)
\end{aligned}
\end{equation*}

y como las variables  son independientes forman una densidad conjunta:

\begin{equation*}
\begin{aligned}
f_{X_1,X_2}(x_1,x_2)=\frac{1}{900}I_{(0,30)}(x_1) I_{(0,30)}(x_2)
\end{aligned}
\end{equation*}



Deseamos encontrar la probabilidad de que la variable aleatoria que representa la diferencia en los tiempos de llegada sea menor a 10. Dado que uno de las dos personas puede llegar en primer lugar o en segundo lugar, tomamos la diferencia en valor obsoluto
$$D= |X_1 - X_2|$$
$$¿P\{D \leq 10\}?$$

Observe que 
\begin{align*}
D\leq10 && \Rightarrow && |x_1 - x_2|\leq10 && \Rightarrow && \left\{ \begin{array}{lcc} x_2 - x_1 \leq10 \\ x_1 - x_2 \leq10  \end{array} \right.
\end{align*}

El conjunto de puntos donde se cumple la desigualdad anterior se puede representar gráficamente.

Necesitamos representar en el plano las siguientes rectas. 

$\left\{ \begin{array}{lcc} x_2 - x_1 \leq 10 \\ x_1 - x_2 \leq 10  \end{array} \right.$

Ahora ubicamos una coordenada para la primera recta el cual sera $(0,30)$ y evaluamos 

\begin{gather*}
     30-0 &\leq 10\\
     30 &\leq 10
 \end{gather*}
 
 dado que no se cumple la desigualdad, entonces nos interesa el punto por debajo de la recta $x_2 - x_1 \leq 10$.
 
 Ahora ubicamos una coordenada por la siguiente recta la cual sera $(30,0)$ y evaluamos 
 
\begin{gather*}
     30-0 &\leq 10\\
     30 &\leq 10
 \end{gather*}
 
 dado que no se cumple la desigualdad, entonces nos interesa el punto por encima de la recta $x_1 - x_2 \leq 10$.
 
 Esta región de interés se ve representado en la siguiente imagen

.


Ahora nos interesa saber el área de la región sombreada, encerrado por el cuadrado $[0,30] \times [0,30]$, observemos podemos obtener esta área de forma sencilla de forma geométrica.

\begin{gather*}
Área_{sombreada}=Área_{cuadrado}-2 * Área_{triángulo} \\
Área_{sombreada}=(30)(30)-2 *  \frac{(20)(20)}{2} \\
Área_{sombreada}=900-400 \\
Área_{sombreada}=500 \\
\end{gather*}

Finalmente, solo queda calcular la probabilidad de que el tiempo que una persona espere a la otra sea como máximo $10$ min, como el $Área_{sombreada} * f_{X_1,X_2}$

\begin{equation*}
\begin{aligned}
P\{D \leq 10\} &= Área_{sombreada} * f_{X_1,X_2}(x_1,x_2)\\
P\{D \leq 10\}&= 500\left(\frac{1}{900}\right)\\
P\{D \leq 10\}&= \frac{5}{9}
\end{aligned}
\end{equation*}

**PASO 7**. Obtenga la gráfica de estabilización que indica que el tamaño de muestra utilizado es suficiente para
garantizar la convergencia del resultado.

```python
import numpy as np
import random
import matplotlib.pyplot as plt
import math
```

```python
#Esta función representa un único experimento. Simula la llegada de las 2 personas y calcula la diferencia.
#Regresa el valor de la diferencia
def diferencia():
    x1=random.uniform(0,30)
    x2=random.uniform(0,30)
    dif=abs(x1-x2)
    return dif
```

```python
#Esta función realiza "n" veces el experimento o ensayo.
def experimento(iteraciones):
    i = 1 # contador del ensayo
    favorables=0 #contador de exitos
    Lista = [] #en esta lista se guarda la aproximación de la prob. en cada iteración.
    
    #El siguiente ciclo se encarga de realizar las n iteraciones
    while i <= iteraciones:
        #condición de "exito". Si la diferencia de tiempos de llegada es menor a 10, se considera un éxito
        if diferencia()<10:
            favorables+=1
        aprox_prob = favorables/i
        Lista.append(aprox_prob)
        i+=1
    return Lista
```

```python
exp = experimento(2000)
```

```python
np.mean(exp)
```

```python
plt.plot(exp)
plt.xlabel('iteraciones')
plt.ylabel('p')
plt.title('Probabilidad de obtener 2 caras')
plt.show()
```

**PASO 10**. Halle el intervalo de confianza de la probabilidad de queaparezcan dos caras lanzando tres veces una moneda
al aire con un nivel de confianza del $95 \%$.

\begin{equation*}
\overline{x}- Z_{\frac{\alpha}{2}}  \cdot \frac{S}{\sqrt{N}} < \mu < \overline{x}+ Z_{\frac{\alpha}{2}}  \cdot \frac{S}{\sqrt{N}}
\end{equation*}

```python
from scipy.stats import norm #Importamos la libreria para poder obtener el valor de Z.
alpha = 1 - 0.95 #Alpha es igual a 1 - la confianza del intervalo
z_alpha_2 = norm.ppf(1-alpha / 2) #Mandamos a pedir el valor de Z para \alpha/2, sabemos que si es el 95% de confienza, Z esta cercano al 1.96.
l_inferior = media - z_alpha_2 * (desv / math.sqrt(N)) #Sacamos limite inferior por la formula
l_superior = media + z_alpha_2 * (desv / math.sqrt(N)) #Sacamos limite superior por la formula
print("Entonces: ", l_inferior, "< \u03BC <", l_superior, "\nCon un 95% de confianza") #El intervalo en donde se encuentra el valor real a una confienza del 95%
```
