# 3) Escriba un programa en Python con el m√©todo de **Acierto y error**

```python
import math
import random
import matplotlib.pyplot as plt

#EJEMPLO DE FUNCI√ìN g(x) DE LA CU√ÅL QUEREMOS SABER SU INTEGRAL
def funcion(x):
    return math.sqrt(1-x**2)



#a,b controlan en intervalo (a,b) sobre el eje "X" sobre el cual se quiero integrar,
#c es la altura del rect√°ngulo en el plano XY que se tomar√° como espacio muestral "omega"
a=0
b=1
c = 2


n=2000
aciertos = 0

#Las siguientes listas nos ayudan a guardar los datos que vamos generando en cada iteraci√≥n para
#depues poder graficar
area=[]
ensayos=[]

#El siguiente ciclo representa cada iteraci√≥n del algoritmo
for i in range(1,n+1):
    #1. Generamos una secuencia de 2N numeros aleatorios
    #2. Organizamos los n√∫meros aleatorios en N pares
    u1=(random.random())
    u2=(random.random())

    #3. Calculamos una uniforme U(a,b) apartir de U(0,1), as√≠ como g(X_i)
    x = (b-a)*u1+a
    g_x = funcion(x)

    #4. Contamos el n√∫mero de casos de acierto N_H
    if g_x > c*u2:
      aciertos += 1

    #Actualizamos las listas que nos sirven para graficar
    ensayos.append(i)
    area.append(c*(b-a)*aciertos*4/i)
#5. Estimamos la integral I como c(b-a)(N_H / N)
estimacion = c*(b-a)*(aciertos/n)
pi = estimacion*4
print(pi)
plt.plot(ensayos,area)
plt.xlabel('N√∫mero de ensayos')
plt.ylabel('Aproximaci√≥n al valor de $\pi$')
plt.show()
```

# 4) Aproximaci√≥n al valor de $\pi$

```python
import math
import random
import matplotlib.pyplot as plt
n=10000
area=[]
ensayos=[]
aciertos=0
for i in range(1,n):
    x1=(random.random())
    x2=(random.random())
    if x1**2+x2**2<1:
        aciertos+=1

    ensayos.append(i)
    area.append(aciertos*4/i)


pi = (aciertos/n)*4
print(pi)
plt.plot(ensayos,area)
plt.xlabel('N√∫mero de ensayos')
plt.ylabel('Aproximaci√≥n al valor de $\pi$')
plt.show()
```

# Aproximaci√≥n a la integral por Monte Carlo Acierto y Error
\begin{equation*}
\begin{aligned}
 I = \int_0^1 \sqrt{arctan(x)} \quad dx
\end{aligned}
\end{equation*}

```python
import math
import random
import matplotlib.pyplot as plt
def funcion(x):
    return math.sqrt(math.atan(x))
n=2000
a=0
b=1
c = 2
aciertos = 0
area=[]
ensayos=[]
for i in range(1,n+1):
    u1=(random.random())
    u2=(random.random())
    x = (b-a)*u1+a
    g_x = funcion(x)
    if g_x > c*u2:
      aciertos += 1
    ensayos.append(i)
    area.append(c*(b-a)*aciertos/i)

estimacion = c*(b-a)*(aciertos/n)
print(estimacion)
plt.plot(ensayos,area)
plt.xlabel('N√∫mero de ensayos')
plt.ylabel('Aproximaci√≥n a la integral de $g(x)$')
plt.show()
```

# 6. Los coeficientes $p$ y $q$ de la ecuaci√≥n cuadr√°tica.
\begin{equation*}
\begin{aligned}
 x^2 + px + q = 0
\end{aligned}
\end{equation*}
# son tomados aleatoriamente en el intervalo $(0,2)$. ¬øCu√°l es la probabilidad de que las raices de esta ecuaci√≥n sean n√∫meros reales?

# a) Resuelve este problema por simulaci√≥n.
Consideremos que para que las raices de esta ecuaci√≥n cuadratica sean reales necesitamso que el discriminante sea de la forma:
\begin{equation*}
\begin{aligned}
\bigtriangleup = b^2 - 4ac \geq 0
\end{aligned}
\end{equation*}
en donde $b=p$ y $c=q$, entonces, para que las raices de esta ecuaci√≥n cuadratica sean reales es necesario que se cumpla:
\begin{equation*}
\begin{aligned}
\bigtriangleup = p^2 - 4q \geq 0
\end{aligned}
\end{equation*}
Consideremos que $p \sim U(0,2)$ y $q \sim U(0,2)$, asumiendo independencia entre estas variables.

Para calcular la probabilidad de que las raices de la funcion cuadratica sean reales, podriamos aproximarlo mediamente una simulaci√≥n. A continuaci√≥n se agrega el codigo, asi como su explicaci√≥n.

```python
import random
import matplotlib.pyplot as plt
```

La cantidad de experimentos que realizaremos seran 1,000,000

```python
N=1000000

#Tendremos un contador que contara la cantidad de veces que se cumple la condici√≥n
#del discriminante positivo

exitos=0
probabilidad_estimada = []
ensayos=[]
```

El siguiente ciclo generara **N=1,000,000** $p$ y $q$ con distribuci√≥n $U(0,2)$, recordemos que la funci√≥n random de python genera aleatorios entre $(0,1)$, por la transformada inversa podemos generar aleatorios que se distribuyen como cualquier Uniforme entre $(a,b)$. Esto por medio de $x = (b-a)u+a$ en donde $u$ es un aleatorio entre 0 y 1. Y el $if$ verifica si la condici√≥n sobre el discriminante se cumple, si esta se cumple, entonces la cantidad de exitos aumentara en 1.

```python
for i in range(1,N+1):
  p = 2*random.random()
  q = 2*random.random()
  if (p**2-4*q)>=0:
    exitos += 1
  ensayos.append(i)
  probabilidad_estimada.append(exitos/i)
```

Entonces, la probabilidad estimada de que las raices sean reales seria la cantidad de exitos entre la cantidad de experimentos, es decir:
\begin{equation*}
\begin{aligned}
\text{probabilidad raices reales} = \frac{exitos}{N}
\end{aligned}
\end{equation*}

```python
plt.plot(ensayos,probabilidad_estimada)
plt.show()

print(probabilidad_estimada[-1])
```

\end{aligned}
\end{equation*}
Consideremos que $p \sim U(0,2)$ y $q \sim U(0,2)$, asumiendo independencia entre estas variables.\\

As√≠:
\begin{equation*}
\begin{aligned}
f_p (p) = \frac{1}{2}I_{(0,2)}(p) && f_Q (q) = \frac{1}{2}I_{(0,2)}(q)
\end{aligned}
\end{equation*}

y como las variables que son independientes:

\begin{equation*}
\begin{aligned}
f_{P,Q}=\frac{1}{4}I_{(0,2)}(p) I_{(0,2)}(q)
\end{aligned}
\end{equation*}

Verificamos que es, en efecto, una densidad.

\begin{equation*}
\begin{aligned}
\int_{0}^{2}\int_{0}^{2}f_{P,Q}(p,q)dpdq&=\int_{0}^{2}\int_{0}^{2}\frac{1}{4}dpdq\\&=\int_{0}^{2}\frac{p}{4}\big|_0^2dq\\&=\int_{0}^{2}\frac{1}{2}dq\\&=\frac{q}{2}\big|_0^2=1
\end{aligned}
\end{equation*}

Geom√©tricamente f_{P,Q} representa un volumen, un prisma cuya base es el rect√°ngulo $[0,2]x[0,2]$ y altura $1/4$.

Vamos a centrar nuestra atenci√≥n en el rect√°ngulo $[0,2]x[0,2]$.

Para la condici√≥n $ùö´\geq 0$ se tiene lo siguiente:

\begin{equation*}
\begin{aligned}
P^2-4Q\geq0&=P^2\geq 4Q
\\&=P\geq 2\sqrt{Q}
\end{aligned}
\end{equation*}

Donde hemos tomado la ra√≠z positiva ya que los valores de Q son todos positivos

Etonces trabajamos con el √°rea por arriba de la par√°bola $P = 2\sqrt{Q}$
